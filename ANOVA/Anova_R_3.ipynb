#Análisis de varianza 
Una prueba estadística clave en los campos de investigación, el Análisis de varianza (ANOVA) es muy útil para analizar conjuntos de datos. Permite realizar comparaciones entre tres o más grupos de datos. Aquí, resumimos las diferencias clave entre estas dos pruebas, incluidas las suposiciones e hipótesis que deben hacerse sobre cada tipo de prueba.

Hay dos tipos de ANOVA que se usan comúnmente, el ANOVA de una vía y el ANOVA de dos vías. Este artículo explorará esta importante prueba estadística y la diferencia entre estos dos tipos de ANOVA.

##¿Qué es un ANOVA unidireccional?

Un ANOVA unidireccional es un tipo de prueba estadística que compara la varianza en las medias grupales dentro de una muestra mientras considera solo una variable o factor independiente. Es una prueba basada en hipótesis, lo que significa que tiene como objetivo evaluar múltiples teorías mutuamente excluyentes sobre nuestros datos. Antes de que podamos generar una hipótesis, necesitamos tener una pregunta sobre nuestros datos para la que queremos una respuesta. Por ejemplo, los investigadores aventureros que estudian una población de morsas podrían preguntar "¿Nuestras morsas pesan más en la temporada de apareamiento temprana o tardía?" Aquí, la variable o factor independiente (los dos términos significan lo mismo) es "mes de la temporada de apareamiento". En un ANOVA, nuestras variables independientes se organizan en grupos categóricos. Por ejemplo, si los investigadores observaran el peso de la morsa en diciembre, enero, febrero y marzo, se analizarían cuatro meses y, por lo tanto, cuatro grupos para el análisis.

Un ANOVA unidireccional compara tres o más de tres grupos categóricos para establecer si hay una diferencia entre ellos. Dentro de cada grupo debe haber tres o más observaciones (aquí, esto significa morsas), y se comparan las medias de las muestras.

###¿Cuáles son las hipótesis de un ANOVA unidireccional?

En un ANOVA unidireccional hay dos hipótesis posibles.

La hipótesis nula (H0) es que no hay diferencia entre los grupos e igualdad entre las medias. (Las morsas pesan lo mismo en diferentes meses)
La hipótesis alternativa (H1) es que hay una diferencia entre las medias y los grupos. (Las morsas tienen diferentes pesos en diferentes meses)


###¿Cuáles son los supuestos de un ANOVA unidireccional?
* Normalidad: que cada muestra se toma de una población normalmente distribuida
* Independencia de la muestra: que cada muestra ha sido extraída
 independientemente de las otras muestras
* Igualdad de varianza: que la varianza de los datos en los diferentes grupos debe ser la misma
* Su variable dependiente, aquí, "peso", debe ser continua, es decir, medida en una escala que se puede subdividir mediante incrementos (es decir, gramos, miligramos).

##¿Qué es un ANOVA de dos vías?

Un ANOVA de dos vías es, como un ANOVA de una vía o ANOVA unidireccional, una prueba basada en hipótesis. Sin embargo, en el ANOVA de dos vías, cada muestra se define de dos maneras y, en consecuencia, se clasifica en dos grupos categóricos. Pensando nuevamente en nuestras morsas, los investigadores podrían usar un ANOVA de dos vías si su pregunta es: "¿Las morsas son más pesadas en la temporada de apareamiento temprana o tardía y eso depende del género de la morsa?" En este ejemplo, tanto el "mes en la temporada de apareamiento" como el "género de la morsa" son factores, lo que significa que en total hay dos factores. Una vez más, se debe considerar el número de grupos de cada factor: para “género” solo habrá dos grupos “masculino” y “femenino”.

Por lo tanto, el ANOVA de dos vías examina el efecto de dos factores (mes y sexo) en una variable dependiente, en este caso el peso, y también examina si los dos factores se afectan entre sí para influir en la variable continua.

##¿Cuáles son los supuestos de un ANOVA de dos vías?
* Su variable dependiente, aquí, "peso", debe ser continua, es decir, medida en una escala que se puede subdividir mediante incrementos (es decir, gramos, miligramos)
* Sus dos variables independientes, aquí, "mes" y "género", deben estar en grupos categóricos e independientes.
* Independencia de la muestra: que cada muestra se ha extraído independientemente de las otras muestras
* Igualdad de varianza: que la varianza de los datos en los diferentes grupos debe ser la misma
* Normalidad: que cada muestra se toma de una población normalmente distribuida

###¿Cuáles son las hipótesis de un ANOVA de dos vías?

Debido a que el ANOVA bidireccional considera el efecto de dos factores categóricos y el efecto de los factores categóricos entre sí, existen tres pares de hipótesis nulas o alternativas para el ANOVA bidireccional. Aquí, los presentamos para nuestro experimento de morsa, donde el mes de la temporada de apareamiento y el género son las dos variables independientes.

* H0: las medias de todos los grupos mensuales son iguales
* H1: la media de al menos un mes de grupo es diferente


* H0: las medias de los grupos de género son iguales
* H1: los medios de los grupos de género son diferentes


* H0: no hay interacción entre el mes y el género
* H1: hay interacción entre el mes y el género


## Resumen: diferencias entre ANOVA unidireccional y bidireccional
Las diferencias clave entre ANOVA unidireccional y bidireccional se resumen claramente a continuación.

1. Un ANOVA unidireccional está diseñado principalmente para permitir la prueba de igualdad entre tres o más medios. Un ANOVA de dos vías está diseñado para evaluar la interrelación de dos variables independientes en una variable dependiente.

2. Un ANOVA unidireccional solo involucra un factor o variable independiente, mientras que hay dos variables independientes en un ANOVA bidireccional.

3. En un ANOVA unidireccional, el único factor o variable independiente analizada tiene tres o más grupos categóricos. Un ANOVA de dos vías en cambio compara múltiples grupos de dos factores.

4. ANOVA unidireccional necesita satisfacer solo dos principios de diseño de experimentos, es decir, replicación y aleatorización. A diferencia del ANOVA de dos vías, que cumple con los tres principios de diseño de experimentos que son la replicación, la aleatorización y el control local.


```r
# Cargar biblioteca
library(tidyverse)
```

```r
# Cargar datos
my_data <- PlantGrowth
```

```r
# Mostrar muestra aleatoria
set.seed (420)
dplyr::sample_n(my_data, 10)
```

En la terminología R, la columna "grupo" se llama factor y las diferentes categorías ("ctr", "trt1", "trt2") se denominan niveles de factor. Los niveles están ordenados alfabéticamente.

```r
# Mostrar los niveles
levels(my_data$group)
```

```r
# Si los niveles no están automáticamente en el orden correcto, pueden ordenarse de la siguiente manera:
my_data$group <- ordered(my_data$group,
                         levels = c("ctrl", "trt1", "trt2"))
group_by(my_data, group) %>%
  summarise(
    count = n(),
    mean = mean(weight, na.rm = TRUE),
    sd = sd(weight, na.rm = TRUE)
  )
```

```r
# box plot
boxplot(weight ~ group, data = my_data,
        xlab = "Treatment", ylab = "Weight",
        frame = FALSE, col = c("#00AFBB", "#E7B800", "#FC4E07"))


```

```r
# instalar biblioteca
install.packages("gplots")
```

```r
# Plotmeans
library("gplots")
plotmeans(weight ~ group, data = my_data, frame = FALSE,
          xlab = "Treatment", ylab = "Weight",
          main="Mean Plot with 95% CI")
```

Calcular prueba ANOVA unidireccional

Queremos saber si hay alguna diferencia significativa entre los pesos promedio de las plantas en las 3 condiciones experimentales.

La función R aov () se puede usar para responder a esta pregunta. La función summary.aov () se utiliza para resumir el modelo de análisis de varianza.


```r
# Hacer ANOVA
res.aov <- aov(weight ~ group, data = my_data)
# Mostrar tabla
summary(res.aov)
```

##Interpretar el resultado de las pruebas ANOVA de una 1 vía

Con base en los resultados de la prueba res.aov y del resumen (res.aov) donde  el valor p es menor que el nivel de significancia 0.05, podemos concluir que existen diferencias significativas entre los grupos resaltados con “*” en el resumen del modelo.

##Comparación múltiple por pares entre las medias de grupos

En la prueba ANOVA de 1 vía, un valor p significativo indica que algunas de las medias del grupo son diferentes, pero no sabemos qué pares de grupos son diferentes.

Es posible realizar múltiples comparaciones por pares para determinar si la diferencia de medias entre pares específicos de grupo es estadísticamente significativa.

Como la prueba ANOVA es significativa, podemos calcular Tukey HSD (Tukey Honest Significant Differences, R function: TukeyHSD ()) para realizar múltiples comparaciones por pares entre las medias de los grupos.

La función TukeyHD () toma el ANOVA ajustado como argumento.



```r
# Prueba Tukey para comparaciones múltiples apareadas
TukeyHSD (res.aov)
```

Las diferentes columnas en TukeyHSD representan el siguiente:

* diff: diferencia entre medias de los dos grupos
* lwr, upr: el punto final inferior y superior del intervalo de confianza al 95% (predeterminado)
* p adj: valor p después del ajuste para las comparaciones múltiples.


###Verificación de los supuestos de ANOVA: ¿validez de la prueba?

La prueba ANOVA supone que, los datos se distribuyen normalmente y la varianza entre los grupos es homogénea. Podemos verificar eso con algunas gráficas de diagnóstico.

Verifique el supuesto de homogeneidad de varianza

La gráfica de residuos versus ajustes se puede usar para verificar la homogeneidad de las variaciones.

[Residuos](https://statisticsbyjim.com/glossary/residuals/): diferencia entre valores observados y la media de cada observación 


```r
plot(res.aov, 1)
```

```r
aov_residuals <- residuals(object = res.aov)
shapiro.test(x = aov_residuals )
```

# ANOVA 2 Vías

Aquí, usaremos el conjunto de datos R incorporado llamado Crecimiento de diente (). Contiene datos de un estudio que evalúa el efecto de la vitamina C en el crecimiento dental en cobayas (guinea pigs). El experimento se realizó en 60 cobayas, donde cada animal recibió uno de los tres niveles de dosis de vitamina C (0.5, 1 y 2 mg / día) mediante uno de los dos métodos de entrega (jugo de naranja o ácido ascórbico (una forma de vitamina C y codificado como VC) Se midió la longitud del diente y a continuación se muestra una muestra de los datos.


```r

library(tidyverse)
# Almacenar los datos en la variable my_data
my_data <- ToothGrowth
```



Para tener una idea de cómo se ven los datos, mostramos una muestra aleatoria de los datos utilizando la función sample_n ()


```r
# Mostrar una muestra aleatoria
set.seed(667)
dplyr::sample_n (my_data, 10)
```

```r
# Verificar la estructura
str(my_data)
```

De la salida anterior, R considera "dosis" como una variable numérica. Lo convertiremos como una variable de factor (es decir, variable de agrupación) de la siguiente manera.

```r
my_data$dose <- factor (my_data$dose,
                   levels = c(0.5, 1, 2),
                   labels = c("D0.5", "D1", "D2"))
```

```r
# Tablas de Frecuencias
table(my_data$supp, my_data$dose)
```

```r
# Box plot 
boxplot(len ~ supp * dose, data=my_data, frame = FALSE, 
        col = c("#00AFBB", "#E7B800"), ylab="Tooth Length")
```

```r
# Two-way interaction plot
interaction.plot(x.factor = my_data$dose, trace.factor = my_data$supp, 
                 response = my_data$len, fun = mean, 
                 type = "b", legend = TRUE, 
                 xlab = "Dose", ylab="Tooth Length",
                 pch=c(1,19), col = c("#00AFBB", "#E7B800"))
```

###Calcular la prueba ANOVA de dos vías
Queremos saber si la longitud del diente depende del suministro y la dosis.

La función R aov () se puede usar para responder esta pregunta. La función summary.aov () se utiliza para resumir el modelo de análisis de varianza.


```r
res.aov2 <- aov(len ~ supp + dose, data = my_data)
summary(res.aov2)
```

```r
res.aov3 <- aov(len ~ supp + dose + supp:dose, data = my_data)
# sumamry de resultados de ANOVA
summary(res.aov3)
```

Calcular algunas estadísticas resumidas
Calcule la media y la desviación estándar por grupos usando dplyr R dentro de tidyverse:


```r
group_by(my_data, supp, dose) %>%
  summarise(
    count = n(),
    mean = mean(len, na.rm = TRUE),
    sd = sd(len, na.rm = TRUE)
  )
```

Comparación múltiple por pares entre las medias de grupos

En la prueba ANOVA, un valor p significativo indica que algunas de las medias del grupo son diferentes, pero no sabemos qué pares de grupos son diferentes.

Es posible realizar múltiples comparaciones por pares para determinar si la diferencia de medias entre pares específicos de grupo es estadísticamente significativa.


** No necesitamos realizar la prueba para la variable "supp" porque tiene solo dos niveles, que ya han demostrado ser significativamente diferentes por la prueba ANOVA. Por lo tanto, la prueba Tukey HSD se realizará solo para el factor variable "dose". **

```r
TukeyHSD(res.aov3, which = "dose")
```

Verifique los supuestos de ANOVA: ¿validez de la prueba?

ANOVA supone que los datos se distribuyen normalmente y la varianza entre los grupos es homogénea. Podemos verificar eso con algunas gráficas de diagnóstico.

Verifique el supuesto de homogeneidad de varianza
La gráfica de residuos versus ajustes se utiliza para verificar la homogeneidad de las variaciones. En el gráfico a continuación, no hay relaciones evidentes entre los residuos y los valores ajustados (la media de cada grupo), lo cual es bueno. Entonces, podemos asumir la homogeneidad de las variaciones.
